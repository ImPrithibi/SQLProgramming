**QUESTION 1**

**Problem**:  Write a query against the dbo.Orders table that computes for each customer order, both a rank and a dense rank,  partitioned by custid, ordered by qty

**Tables:** 

dbo.Orders

dbo.OrderDetails

**Columns:** 

Orders

> Custid
> 
> Orderid

OrderDetails

> Orderid
> 
> qty

**Predicate:** 

Join Condition: o.orderid = od.orderid. This condition is used to join dbo.Orders and dbo.OrderDetails on their common orderid column, ensuring that the details associated with each order are correctly matched.

The Rank() orders ows with identical values in the order by column receive the same rank, but the next rank is incremented by the total number of tied rows, creating gaps. While DENSE\_RANK() functions similarly by partitioning and ordering rows, but it assigns ranks without creating gaps. 

**Process:** 

This SQL code retrieves information from two tables, Orders and OrderDetails, by joining them on their common orderid field. For each customer (custid), it calculates the total quantity (qty) of items ordered per order, and then assigns two types of rankings based on this total quantity: RANK() gives a unique rank within each customer group, allowing for ties (where the next rank after a tie is incremented by the number of tied ranks), and DENSE\_RANK() does the same without leaving gaps between ranks after ties. The results are grouped by customer and order, then ordered by customer and the total quantity of items in ascending order.

````sql
USE TSQLV4; 
SELECT
    o.custid,
    o.orderid,
    SUM(od.qty) AS qty,
    RANK() OVER (PARTITION BY o.custid ORDER BY SUM(od.qty)) AS rnk,
    DENSE_RANK() OVER (PARTITION BY o.custid ORDER BY SUM(od.qty)) AS drnk
FROM
    dbo.Orders o
JOIN
    dbo.OrderDetails od ON o.orderid = od.orderid
GROUP BY
    o.custid,
    o.orderid
ORDER BY
    o.custid,
    SUM(od.qty);


--query converted 
USE Northwinds2022TSQLV7;
SELECT
    o.CustomerId,
    o.orderid,
    SUM(od.Quantity) AS qty,
    RANK() OVER (PARTITION BY o.CustomerID ORDER BY SUM(od.Quantity)) AS rnk,
    DENSE_RANK() OVER (PARTITION BY o.CustomerID ORDER BY SUM(od.Quantity)) AS drnk
FROM
    Sales.[Order] o
JOIN
    Sales.OrderDetail od ON o.orderid = od.orderid
GROUP BY
    o.CustomerId,
    o.orderid
ORDER BY
    o.CustomerId,
    SUM(od.Quantity);

````

> empid

**Predicate:** 

Partitioning: The data is partitioned by empid, meaning the running total calculation starts over for each new employee ID. 

Ordering: Within each partition (employee), the data is ordered by ordermonth. This ordering dictates how the running total is computed, ensuring it accumulates in the sequence of months. 

Window Frame: The window frame specified by ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW means that for each row, the running total includes all rows from the start of the partition up to the current row.

````sql
---------------------------------------------------------------------
-- Window Functions, Described
---------------------------------------------------------------------

USE TSQLV4;

SELECT empid, ordermonth, val,
  SUM(val) OVER(PARTITION BY empid
                ORDER BY ordermonth
                ROWS BETWEEN UNBOUNDED PRECEDING
                         AND CURRENT ROW) AS runval
FROM Sales.EmpOrders;

--NORTHWINDS 

USE Northwinds2022TSQLV7; 
SELECT EmployeeId, OrderDate, OrderId,
  SUM(OrderId) OVER(PARTITION BY EmployeeId
                ORDER BY orderDate
                ROWS BETWEEN UNBOUNDED PRECEDING
                         AND CURRENT ROW) AS runval
FROM Sales.[Order];
````

**Problem** **How can I view the previous and next values of a specific column for each customer's order, ordered by the date and ID of the order, to analyze trends or changes in order values over time**

**Tables:** 

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">Sales.OrderValues (view)</span>

**Columns:** 

OrderValues

> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">custid</span>
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">orderid</span>
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">val</span>
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">preval(derived)</span>
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">nextval(derived)</span>

**Predicate**

**prevval: A computed column that uses the LAG window function to get the previous val within the partition defined by custid, ordered by orderdate and orderid.** 

**nextval: A computed column that uses the LEAD window function to get the next val within the same partition and ordering criteria.**

````sql
---------------------------------------------------------------------
-- Offset Window Functions
---------------------------------------------------------------------
USE TSQLV4; 
-- LAG and LEAD
SELECT custid, orderid, val,
  LAG(val)  OVER(PARTITION BY custid
                 ORDER BY orderdate, orderid) AS prevval,
  LEAD(val) OVER(PARTITION BY custid
                 ORDER BY orderdate, orderid) AS nextval
FROM Sales.OrderValues
ORDER BY custid, orderdate, orderid;

--NORTHWINDS 
USE Northwinds2022TSQLV7; 
SELECT customerID, orderid,
  LAG(OrderId)  OVER(PARTITION BY custid
                 ORDER BY orderdate, orderid) AS prevval,
  LEAD(OrderId) OVER(PARTITION BY custid
                 ORDER BY orderdate, orderid) AS nextval
FROM Sales.[Order]
ORDER BY CustomerId, orderdate, orderid;
````

**Problem** **How can I find the initial and final values of a certain measure (val) for each customer's sequence of orders, ordered by date and ID, to compare the start and end points of their ordering history?**

**Tables:** 

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">Sales.OrderValues (view)</span>

**Columns:** 

OrderValues

> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">custid</span>
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">orderid</span>
> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">val</span>

**Predicate**

First Value This function is used to obtain the first value in a window or set of rows defined by the OVER clause

The window frame is specified by ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW. This tells SQL to consider all rows from the start of the partition up to the current row (inclusive) when determining the "first value".  

Last Value This function fetches the last value within a specified window or set of rows.   

The specification ROWS BETWEEN CURRENT ROW AND UNBOUNDED FOLLOWING sets the window frame to consider all rows from the current one to the last row in the partition. This ensures that the lastval column for any row reflects the value of val from the last row in the ordered partition.

````sql
USE TSQLV4;
-- FIRST_VALUE and LAST_VALUE
SELECT custid, orderid, val,
  FIRST_VALUE(val) OVER(PARTITION BY custid
                        ORDER BY orderdate, orderid
                        ROWS BETWEEN UNBOUNDED PRECEDING
                                 AND CURRENT ROW) AS firstval,
  LAST_VALUE(val)  OVER(PARTITION BY custid
                        ORDER BY orderdate, orderid
                        ROWS BETWEEN CURRENT ROW
                                 AND UNBOUNDED FOLLOWING) AS lastval
FROM Sales.OrderValues
ORDER BY custid, orderdate, orderid;


--NORTHWINDS 
USE Northwinds2022TSQLV7; 
-- FIRST_VALUE and LAST_VALUE
SELECT CustomerId, orderid,
  FIRST_VALUE(OrderId) OVER(PARTITION BY customerID
                        ORDER BY orderdate, orderid
                        ROWS BETWEEN UNBOUNDED PRECEDING
                                 AND CURRENT ROW) AS firstval,
  LAST_VALUE(OrderId)  OVER(PARTITION BY customerID
                        ORDER BY orderdate, orderid
                        ROWS BETWEEN CURRENT ROW
                                 AND UNBOUNDED FOLLOWING) AS lastval
FROM Sales.[Order]
ORDER BY CustomerId, orderdate, orderid;
````

This SQL code calculates the total value of all orders in the dataset and the total value of orders for each individual customer. It does this by selecting each order's ID, customer ID, and value, and then uses window functions to compute two sums: one for the total value of all orders (totalvalue) without partitioning, applying to the entire dataset, and another for the total value of orders by each customer (custtotalvalue), partitioned by custid.

````sql
USE TSQLV4;
---------------------------------------------------------------------
-- Aggregate Window Functions
---------------------------------------------------------------------

SELECT orderid, custid, val,
  SUM(val) OVER() AS totalvalue,
  SUM(val) OVER(PARTITION BY custid) AS custtotalvalue
FROM Sales.OrderValues;


--NORTHWINDS 
USE Northwinds2022TSQLV7;
---------------------------------------------------------------------
-- Aggregate Window Functions
---------------------------------------------------------------------

SELECT orderid, CustomerId, 
  SUM(OrderId) OVER() AS totalvalue,
  SUM(OrderId) OVER(PARTITION BY customerID) AS custtotalvalue
FROM Sales.[Order];
````

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">Sales.OrderValues (view)</span>

**Columns:**

> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">custid</span>

> 
> <span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">val</span>

**Predicate**

This SQL code computes the percentage contribution of each individual order value (val) to the total order value across all customers and separately within each customer's total order value. It achieves this by dividing the value of each order by the total sum of order values in the dataset (pctall) and by the sum of order values for each customer (pctcust), multiplying each by 100 to express the result as a percentage.

````sql
USE TSQLV4;
SELECT orderid, custid, val,
  100. * val / SUM(val) OVER() AS pctall,
  100. * val / SUM(val) OVER(PARTITION BY custid) AS pctcust
FROM Sales.OrderValues;

--Northwinds 

USE Northwinds2022TSQLV7;
SELECT orderid, CustomerId, 
  100. * OrderId / SUM(OrderId) OVER() AS pctall,
  100. * OrderId / SUM(OrderId) OVER(PARTITION BY customerID) AS pctcust
FROM Sales.[Order];
````
