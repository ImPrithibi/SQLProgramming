\-- customer order:

\-- \* the difference between the current order quantity

\--   and the customer's previous order quantity

\--   and the customer's next order quantity.

  

\-- Desired output:

\------ ----------- ----------- ----------- -----------

C      20002       20          6           NULL

D      30007       30          NULL        NULL

```sql
USE TSQLV4;
SELECT
    o.custid,
    o.orderid,
    od.qty,
    -- Calculate the difference between the current order quantity
    -- and the previous order's quantity for the same customer
    od.qty - LAG(od.qty) OVER(PARTITION BY o.custid ORDER BY o.orderid) AS diffprev,
    -- Calculate the difference between the current order quantity
    -- and the next order's quantity for the same customer
    LEAD(od.qty) OVER(PARTITION BY o.custid ORDER BY o.orderid) - od.qty AS diffnext
FROM dbo.Orders AS o
JOIN dbo.Orderdetails AS od ON o.orderid = od.orderid
ORDER BY o.custid, o.orderid;

USE Northwinds2022TSQLV7;
SELECT
    o.CustomerId,
    o.OrderId,
    od.Quantity,
    -- Calculate the difference between the current order quantity
    -- and the previous order's quantity for the same customer
    od.Quantity - LAG(od.Quantity) OVER(PARTITION BY o.CustomerId ORDER BY o.OrderId) AS diffprev,
    -- Calculate the difference between the current order quantity
    -- and the next order's quantity for the same customer
    LEAD(od.Quantity) OVER(PARTITION BY o.CustomerId ORDER BY o.OrderId) - od.Quantity AS diffnext
FROM Sales.[Order] AS o
JOIN Sales.OrderDetail AS od ON o.OrderId = od.OrderId
ORDER BY o.CustomerId, o.OrderId;
```

\---------------------------------------------------------------------

**Proposition:** Write a query to retrieve the order quantities from dbo.Orders table into different customers by each employee

**Tables:** 

- dbo.Orders

**Columns:**

- empid
    
- A, B, C, D: These are aliases for the summed quantities of orders for each respective customer ID
    

**Predicate:**

- The SELECT statement identified the empid to group the results by each employee
- The total sum is calculated for each customer
- THE CASE within the SUM function checks the custid for each order. If theres no match, then the result is NULL
- The GROUP BY using empid makes sure that the results are aggreagated by the employee and the column shows how much the employee has sold to each customer type

```sql
-- Query against Orders, grouping by employee, pivoting customers,
-- aggregating sum of quantity
USE TSQLV4;
SELECT empid,
  SUM(CASE WHEN custid = 'A' THEN qty END) AS A,
  SUM(CASE WHEN custid = 'B' THEN qty END) AS B,
  SUM(CASE WHEN custid = 'C' THEN qty END) AS C,
  SUM(CASE WHEN custid = 'D' THEN qty END) AS D  
FROM dbo.Orders
GROUP BY empid;


USE Northwinds2022TSQLV7;
SELECT Employeeid,
  SUM(CASE WHEN Customerid = 'A' THEN Quantity END) AS A,
  SUM(CASE WHEN Customerid = 'B' THEN Quantity END) AS B,
  SUM(CASE WHEN Customerid = 'C' THEN Quantity END) AS C,
  SUM(CASE WHEN Customerid = 'D' THEN Quantity END) AS D  
FROM dbo.[Orders]
GROUP BY Employeeid;
```

**Proposition:** Write a query to organize the data from dbo.Orders grouped by the custid for each employee and make the custid for each empid and aggregate the total for each custid under the respective employee

**Tables:**

- dbo.Orders

**Columns:**

**Predicate:**

- The FROM SELECT empid, custid, qty FROM dbo.Orders selects only the relevant columns from the dbo.Orders table
- The PIVOT SUM(qty) transforms the custid values from rows into columns and aggregates the SUMs based on the custid
- The pivot table then retrieves the pivoted columns and shows the total quantity of orders that the employee processed for each customer type

```sql
-- Logical equivalent of previous query using the native PIVOT operator
USE TSQLV4;
SELECT empid, A, B, C, D
FROM (SELECT empid, custid, qty
      FROM dbo.Orders) AS D
  PIVOT(SUM(qty) FOR custid IN(A, B, C, D)) AS P;


USE Northwinds2022TSQLV7;
SELECT Employeeid, A, B, C, D
FROM (SELECT Employeeid, Customerid, Quantity
      FROM dbo.Orders) AS D
  PIVOT(SUM(Quantity) FOR Customerid IN(A, B, C, D)) AS P;
```

**Predicate:**

- Uses the CASE statement within a SUM function to aggregate on the conditions of qty valued based on the custid
- The CASE statement checks each value of the cust id and if it matches then it is included in the column, if not, then it will be NULL

```sql
-- Query demonstrating the problem with implicit grouping
USE TSQLV4;
SELECT empid, A, B, C, D
FROM dbo.Orders
  PIVOT(SUM(qty) FOR custid IN(A, B, C, D)) AS P
-- Logical equivalent of previous query
SELECT empid,
  SUM(CASE WHEN custid = 'A' THEN qty END) AS A,
  SUM(CASE WHEN custid = 'B' THEN qty END) AS B,
  SUM(CASE WHEN custid = 'C' THEN qty END) AS C,
  SUM(CASE WHEN custid = 'D' THEN qty END) AS D  
FROM dbo.Orders
GROUP BY orderid, orderdate, empid;


USE Northwinds2022TSQLV7;
SELECT EmployeeId, A, B, C, D
FROM dbo.Orders
  PIVOT(SUM(Quantity) FOR Customerid IN(A, B, C, D)) AS P
-- Logical equivalent of previous query
SELECT Employeeid,
  SUM(CASE WHEN Customerid = 'A' THEN Quantity END) AS A,
  SUM(CASE WHEN Customerid = 'B' THEN Quantity END) AS B,
  SUM(CASE WHEN Customerid = 'C' THEN Quantity END) AS C,
  SUM(CASE WHEN Customerid = 'D' THEN Quantity END) AS D  
FROM dbo.Orders
GROUP BY Orderid, Orderdate, EmployeeId;
```
